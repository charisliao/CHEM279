# Welcome! 

We will investigate the truncation error between the actual analytical forces and the approximated forces. There are two approaches used for approximation. Therefore, we are going to investigate the errors produced using different step sizes. 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

### For test case 1 

```python

# Specify the path to the file
file_path = "Data/truncation_error.txt"

# Initialize lists to store data
step_sizes = []
forward_truncation_errors = []
central_truncation_errors = []

# Read data from the file
with open(file_path, 'r') as file:
    lines = file.readlines()
    is_forward = False
    is_central = False
    for line in lines:
        line = line.strip()
        if not step_sizes:
            # Read the first line into step_sizes
            step_sizes = [float(value) for value in line.split(', ')]
        elif line.startswith("Truncation error for forward difference"):
            is_forward = True
            is_central = False
        elif line.startswith("Truncation error for central difference"):
            is_forward = False
            is_central = True
        elif line and line[0].isdigit():
            # Remove the trailing comma and split by ', '
            values = [float(value) for value in line.rstrip(',').split(', ')]
            if is_forward:
                forward_truncation_errors.extend(values)
            elif is_central:
                central_truncation_errors.extend(values)

# Now, you have the step sizes, forward truncation errors, and central truncation errors in their respective lists.
print("Step Sizes:", step_sizes)
print("Forward Truncation Errors:", forward_truncation_errors)
print("Central Truncation Errors:", central_truncation_errors)
```

```python
log_step_sizes = np.log(step_sizes)
log_forward_truncation_errors = np.log(forward_truncation_errors)
log_central_truncation_errors = np.log(central_truncation_errors)
```

```python
# Create a plot
plt.figure(figsize=(8, 6))
plt.plot(log_step_sizes, log_forward_truncation_errors, marker='o', label='Forward Truncation Error')
plt.plot(log_step_sizes, log_central_truncation_errors, marker='s', label='Central Truncation Error')

# Set plot labels and title
plt.xlabel('Log Step Sizes')
plt.ylabel('Log Truncation Error')
plt.title('Log Truncation Error vs. Log Step Sizes')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
